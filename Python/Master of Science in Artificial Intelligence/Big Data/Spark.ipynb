**Create a RDD from an existing list**

```python
names_list = ["al", "ani", "jackie", "lalitha", "mark", "neil", "nick", "laurence", "shirin", "jackie", "mark", "ani", "mark"]
names_rdd = sc.parallelize(names_list)
print("Done!")
```

**Use of the transformation map()**

```python
names_rdd = names_rdd.map(lambda name: name.capitalize())
names_rdd.collect()
```

**Use of transformation filter()**

```python
l_names_rdd = names_rdd.filter(lambda name: name[0] =="L")
l_names_rdd.collect()
```

**Use of transformation distinct()**

```python
names_rdd = names_rdd.distinct()
names_rdd.collect()
```

**Use of transformation union()**

```python
union_rdd = names_rdd.union(l_names_rdd)
union_rdd.collect()
```

**Use of transformation intersection()**

```python
intersect_rdd = names_rdd.intersection(l_names_rdd)
intersect_rdd.collect()
```

**Use of transformation subtract()**

```python
subtract_rdd = names_rdd.subtract(l_names_rdd)
subtract_rdd.collect()
```

**Use of transformation cartesian()**

```python
cartesian_rdd = names_rdd.cartesian(l_names_rdd)
cartesian_rdd.collect()
```

**Use of the action count()**

```python
names_rdd.count()
```

**Use of the action first()**

```python
names_rdd.first()
```

**Use of the action take()**


```python
names_rdd.take(5)
```

**Use of the action takeSample()**

```python
first_sample = names_rdd.takeSample(False, 3)
second_sample = names_rdd.takeSample(True, 3)
third_sample = names_rdd.takeSample(False, 3, 2)

print(first_sample)
print(second_sample)
print(third_sample)
```

**Create a RDD from an external source**

```python
data_file = "./data/neo4j.log"
lines_rdd = sc.textFile(data_file)
lines_rdd.take(5)
```

**Use of the transformation flatMap()**

```python
lines_rdd_map = lines_rdd.map(lambda line: line.split())
lines_rdd_map.take(5)
```

```python
lines_rdd_flatmap = lines_rdd.flatMap(lambda line: line.split())
lines_rdd_flatmap.take(5)
```

**Use of the action reduce()**

```python
no_words_per_line = lines_rdd_map.map(lambda x: len(x))
no_words_per_line.collect()
```

```python
no_words = no_words_per_line.reduce(lambda x, y: x + y)
print(no_words)
```

**Use of the action aggregate()**

```python
no_words = lines_rdd_map.aggregate(0, (lambda acc, value:  acc + len(value)), (lambda x, y: x+y))
print(no_words)
```

**Use of the action countByValue()**

```python
occurrences = lines_rdd_flatmap.countByValue()
print(occurrences)
```

**Use of the transformation groupByKey()**

```python
word_occ_rdd_gk = word_occ_rdd.groupByKey()
word_occ_rdd_gk.take(5)
```

```python
word_occ_rdd_gk_list = word_occ_rdd_gk.map(lambda x: (x[0], list(x[1])))
word_occ_rdd_gk_list.take(5)
```

**Use of tranformation mapValues()**

```python
word_occ_rdd_gk_list = word_occ_rdd_gk.mapValues(lambda x: list(x))
word_occ_rdd_gk_list.take(5)
```

**Use of the transformation reduceByKey()**

```python
word_occ_rdd = word_occ_rdd.reduceByKey(lambda x, y : x + y)
word_occ_rdd.take(5)
```

**Use of the transformation sortBy()**


```python
word_occ_rdd = word_occ_rdd.sortBy(lambda x:x[1], ascending=False)
word_occ_rdd.take(5)
```

**Use of the transformation cogroup()**

```python
r1 = sc.parallelize([("Panda", 3), ("Pink", 2), ("Pink", 5), ("Red", 10)])
r2 = sc.parallelize([("Panda", 1), ("Pink", 7), ("Pirate", 7)])

res = r1.cogroup(r2)
res.collect()
```

```python
res = res.mapValues(lambda x: (list(x[0]), list(x[1])))
res.collect()
```

**Use of the transformation join()**

```python
res = r1.join(r2)
res.collect()
```

**Use of the transformation leftOuterJoin()**

```python
res = r1.leftOuterJoin(r2)
res.collect()
```

**Use of the transformation rightOuterJoin()**

```python
res = r1.rightOuterJoin(r2)
res.collect()
```
